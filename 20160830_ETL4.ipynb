```python
# Open database connection
db = MySQLdb.connect("localhost","root", db="testdb")

# prepare a cursor object using cursor() method
cursor = db.cursor()
# Drop table if it already exist using execute() method.
cursor.execute("DROP TABLE IF EXISTS CONTACTS")

# Create table as per requirement
sql = """CREATE TABLE CONTACTS (
         NAME  VARCHAR(255) NOT NULL,
         JOB  VARCHAR(255),
         HOBBY VARCHAR(255))"""

cursor.execute(sql)

# disconnect from server
db.close()
```

```python
# Open database connection
db = MySQLdb.connect("localhost","root", db="testdb")

myDict = {'name': 'ian', 'job': 'engineer', 'hobby': 'cycling'}

# prepare a cursor object using cursor() method
cursor = db.cursor()
placeholders = ', '.join(['%s'] * len(myDict))
columns = ', '.join(myDict.keys())
sql = "INSERT INTO %s ( %s ) VALUES ( %s )" % (table, columns, placeholders)
print sql
cursor.execute(sql, myDict.values())
db.commit()
db.close()
```

```python
db = MySQLdb.connect("localhost","root", db="testdb")
mybulklist = [{'name': 'ian', 'job': 'engineer', 'hobby': 'cycling'},
             {'name': 'john', 'job': 'ceo', 'hobby': 'running'},
             {'name': 'amy', 'job': 'designer', 'hobby': 'painting'}]
columns = ', '.join(mybulklist[0].keys())
print columns
cursor = db.cursor()
values_to_insert = [tuple(x.values()) for x in mybulklist]
print values_to_insert
query = "INSERT INTO {} ( {} ) VALUES ".format(table, columns) + ",".join("(%s, %s, %s)" for _ in values_to_insert)
print query
flattened_values = [item for sublist in values_to_insert for item in sublist]
print flattened_values
cursor.execute(query, flattened_values)
db.commit()
db.close()
```

```python
# Open database connection
db = MySQLdb.connect("localhost","root",db="testdb" )

# prepare a cursor object using cursor() method
cursor = db.cursor()

# Prepare SQL query to INSERT a record into the database.
sql = "SELECT * FROM CONTACTS"
try:
    cursor.execute(sql)
    results = cursor.fetchall()
    for row in results:
        name = row[0]
        job = row[1]
        hobby = row[2]

        print "name is {}, job is {}, hobby is {}".format(name, job, hobby)
except:
    print "Error: unable to fecth data"

# disconnect from server
db.close()
```
