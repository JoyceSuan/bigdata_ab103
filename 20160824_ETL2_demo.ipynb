```python
# Object Oriented Programming
#
# class Name(object):
#   def __init__(self, *args):
#       self.
#
#
class Human(object):
    # Class attributes
    species = "monkey"
    
    def __init__(self, name):
        # instance attribute
        self.name = name
        self.age = 0
    
    # instance method
    def say(self, msg):
        return "{}: {}".format(self.name, msg)
    
    @classmethod
    def get_species(cls):
        return cls.species
    
    @staticmethod
    def grunt():
        return "grunt...."
    
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, age):
        self._age = age
        
    def __str__(self):
        return "This is a human class"
    
    
    
        
    
```

```python
ian = Human(name='Ian')
```

```python
# calling a method
ian.say("HI")
```

```python
# setter
ian.age = 30
```

```python
# getter
ian.age
```

```python
print ian
```
